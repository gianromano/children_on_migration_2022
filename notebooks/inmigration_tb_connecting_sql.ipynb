```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
from sqlalchemy import text

import warnings

# Suprimir todos los warnings
warnings.filterwarnings("ignore")

# Suprimir los warnings de una categoría específica
warnings.filterwarnings("ignore", category=DeprecationWarning)

import getpass  # To get the password without showing the input
password = 'root'
```

```python
bd = "inmigration_europe"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+bd
engine = create_engine(connection_string)
```

```python
def inmigration_tb(engine):
    with engine.connect() as connection:
        query = text(f'''SELECT * FROM inmigration_tb''')
        result = connection.execute(query)
        # Fetch all rows from the result
        rows = result.fetchall()
        # Get column names from the result set
        columns = result.keys()
        # Create DataFrame with column names
        df_result = pd.DataFrame(rows, columns=columns)
    return df_result
```

```python
df_result = inmigration_tb(engine)
df_result
```

```python
df_result = df_result.drop(columns='id', axis=1)
df_result
```
