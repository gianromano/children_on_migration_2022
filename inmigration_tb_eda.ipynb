```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Suprimir todos los warnings
warnings.filterwarnings("ignore")

# Suprimir los warnings de una categoría específica
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
df = pd.read_csv("C:/Users/user/Jupyter/IronHack/Work_Python/Final Project/Data-csv-Tableau/df_result_INMIGRATION_CITIZEN.csv", sep=';')
```

```python
df
```

```python
df['age']=df['age'].astype('category')
df['sex']=df['sex'].astype('category')
df['geo']=df['geo'].astype('category')
df['citizen']=df['citizen'].astype('category')
df
```

```python
df.info()
```

```python
# Asumiendo que ya tienes el DataFrame df_result cargado.
# Descripción estadística de 'OBS_VALUE'
print(df['OBS_VALUE'].describe())

# Histograma de 'OBS_VALUE'
plt.figure(figsize=(10, 6))
sns.histplot(df['OBS_VALUE'], bins=20, kde=True)
plt.title('Distribución de los Valores Observados')
plt.xlabel('Valor Observado')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Descripción de 'OBS_VALUE' por 'age'
print(df.groupby('age')['OBS_VALUE'].describe())

# Barplot de 'OBS_VALUE' por 'age'
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='OBS_VALUE', data=df)
plt.title('Distribución de Valores Observados por Edad')
plt.xlabel('Edad')
plt.ylabel('Valor Observado')
plt.xticks(rotation=90)
plt.show()
```

```python
# Descripción de 'OBS_VALUE' por 'sex'
print(df.groupby('sex')['OBS_VALUE'].describe())

# Barplot de 'OBS_VALUE' por 'sex'
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='OBS_VALUE', data=df)
plt.title('Distribución de Valores Observados por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Valor Observado')
plt.show()
```

```python
# Descripción de 'OBS_VALUE' por 'geo'
print(df.groupby('geo')['OBS_VALUE'].describe())

# Barplot de 'OBS_VALUE' por 'geo'
plt.figure(figsize=(10, 6))
sns.barplot(x='geo', y='OBS_VALUE', data=df)
plt.title('Distribución de Valores Observados por Región Geográfica')
plt.xlabel('Región Geográfica')
plt.ylabel('Valor Observado')
plt.xticks(rotation=90)
plt.show()
```

```python
# Descripción de 'OBS_VALUE' por 'TIME_PERIOD'
print(df.groupby('TIME_PERIOD')['OBS_VALUE'].describe())

# Barplot de 'OBS_VALUE' por 'TIME_PERIOD'
plt.figure(figsize=(10, 6))
sns.barplot(x='TIME_PERIOD', y='OBS_VALUE', data=df)
plt.title('Distribución de Valores Observados por Período de Tiempo')
plt.xlabel('Período de Tiempo')
plt.ylabel('Valor Observado')
plt.xticks(rotation=90)
plt.show()
```

```python
# Descripción de 'OBS_VALUE' por 'citizen'
print(df.groupby('citizen')['OBS_VALUE'].describe())

# Dibujar el strip plot
plt.figure(figsize=(14, 8))
sns.stripplot(x='citizen', y='OBS_VALUE', data=df, jitter=True)
plt.title('Distribución de Valores Observados por Ciudadano')
plt.xlabel('Ciudadano')
plt.ylabel('Valor Observado')
plt.xticks(rotation=90)
plt.show()
```
